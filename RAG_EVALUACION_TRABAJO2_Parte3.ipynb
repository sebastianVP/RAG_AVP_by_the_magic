```python
# Recuperaci√≥n de Informaci√≥n usando TF-IDF
def retrieve_document(query, contexts):
    # Vectorizar los contextos y la consulta
    vectorizer = TfidfVectorizer()
    context_vectors = vectorizer.fit_transform(contexts)
    query_vector = vectorizer.transform([query])

    # Calcular similitud coseno
    scores = cosine_similarity(query_vector, context_vectors).flatten()
    best_match_index = scores.argmax()
    return contexts[best_match_index], scores[best_match_index]
```

##**2.Recuperaci√≥n de informaci√≥n usando modelos Hugging Face**

Cargaremos un modelo preentrenado

**Ejemplo: distilbert**

```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModel.from_pretrained("distilbert-base-uncased")


# Lista de modelos
model_names = [
    "distilbert-base-uncased",
    "bert-base-uncased",
    "roberta-base",
    "albert-base-v2",
    "xlnet-base-cased",
    "facebook/bart-base"
]
```

```python
# Definicion DEL METODO DE RECUPERACION
def retrieve_document(query, contexts):
    query_encoding = tokenizer(query, return_tensors="pt", padding=True, truncation=True)
    query_vector = model(**query_encoding).last_hidden_state.mean(dim=1)  # Embedding de la consulta

    context_vectors = []
    for context in contexts:
        context_encoding = tokenizer(context, return_tensors="pt", padding=True, truncation=True)
        context_vector = model(**context_encoding).last_hidden_state.mean(dim=1)
        context_vectors.append(context_vector)

    context_vectors = torch.cat(context_vectors, dim=0)
    scores = torch.mm(query_vector, context_vectors.T).squeeze()  # Producto escalar
    best_match_index = scores.argmax().item()

    return contexts[best_match_index], scores[best_match_index].item()
```
